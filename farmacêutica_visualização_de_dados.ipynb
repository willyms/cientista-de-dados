# **Desafio**

  Diversos gestores têm um certo grau de dificuldade na hora de consolidar e apresentar dados e informações do próprio setor, seja para outra empresa ou para os seus superiores.

  Imagine que você é responsável pela área de vendas de uma grande empresa farmacêutica e precisa criar uma apresentação para os seus superiores sobre os dez medicamentos mais vendidos no último mês, bem como a quantidade das unidades vendidas. São eles:
```
dataset: [
  "Produtos": {"Dorflex","Xarelto","Selozok","Neosaldina","Torsilax","Aradois","Glifage XR", "Addera D3", "Anthelios", "Buscopan composto"},
  "Unidades": {847,674,610,530,502,478,445,310,220,218}
]
```

````python
dataset = [
 ["Dorflex","Xarelto","Selozok","Neosaldina","Torsilax","Aradois","Glifage XR", "Addera D3", "Anthelios", "Buscopan composto"],
 [847,674,610,530,502,478,445,310,220,218]
]
````

01 - importar os dados para o dataframe 

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({"produtos": dataset[0], "unidades": dataset[1]})

df
````

02 - ordenar a coluna 'unudades' ordem maior para o menor

````python
df = df.sort_values(by='unidades', ascending=False)

df
````

03 - Apresentação grafico

- As duas melhores representações gráficas para o problema citado devem ser os de colunas ou barras, pois ambos são utilizados para mostrar componentes de um todo e são mais utilizados quando se tem poucos dados. Neste caso, 10 remédios e suas quantidades vendidas no último mês.
- Pode ser utilizado também, um gráfico pizza para mostrar a porcentagem de venda de cada medicamento.
- Gráficos, como linhas e áreas, devem ser utilizados nos casos de comparação de dados, exemplo: comparação de vendas de um mês para o outro.

Exemplo simples de gráficos com barras, colunas e pizza:

````python
ax = df.plot(kind='bar', y='unidades', figsize=(10,5))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
````

````python
df.plot.pie(y='unidades', title="Grafico Pizza", legend=False, \
        autopct='%1.1f%%', figsize=(7, 7), explode=(0.2,0,0,0,0,0,0,0,0,0), \
        shadow=True, startangle=0)
````
